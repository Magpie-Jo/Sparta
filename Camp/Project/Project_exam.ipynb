```python
# Host 숙소 소개글의 작성길이 파악 위해 글자 수 추출
bnb['host_about'] = bnb['host_about'].fillna('')

host_about_len = bnb['host_about'].str.len()

bnb.insert(11,'host_about_len',host_about_len)
```

```python
# 호스트 가입일 결측값을 최초 리뷰일로 대체
bnb['host_since'] = bnb['host_since'].fillna(bnb['first_review'])

# 최초 리뷰일이 NaN인 관계로 가입일에 적용 불가한 데이터 삭제
bnb = bnb.dropna(subset=['host_since'])

# 삭제 확인
bnb['host_since'].isnull().sum()
```
