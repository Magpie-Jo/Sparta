```python
#Ï¥àÍ∏∞ ÏÑ§Ï†ïÎ∞è ÏãúÏä§ÌÖú ÎùºÏù¥Î∏åÎü¨Î¶¨
import platform
import warnings

# Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî ÎùºÏù¥Î∏åÎü¨Î¶¨
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta

warnings.filterwarnings('ignore')

print(platform.system())
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_colwidth', None)

# ÏãúÍ∞ÅÌôî OSÎ≥Ñ ÌïúÍ∏ÄÌè∞Ìä∏ ÏÑ§Ï†ï
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows Ìè∞Ìä∏ ÏÑ§Ï†ï
elif platform.system() == 'Mac':
    plt.rcParams['font.family'] = 'AppleGothic'  # Mac Ìè∞Ìä∏ ÏÑ§Ï†ï
```

```python
# Ï†ïÏ†ú Îç∞Ïù¥ÌÑ∞ CSV Ï†ÄÏû•  Î∞è Î∂àÎü¨Ïò§Í∏∞

if platform.system() == 'Windows':
    df_ch = pd.read_csv('TFT_Challenger_MatchData.csv')
    df_gm = pd.read_csv('TFT_GrandMaster_MatchData.csv')
    df_ma = pd.read_csv('TFT_Master_MatchData.csv')
    df_di = pd.read_csv('TFT_Diamond_MatchData.csv')
    df_pt = pd.read_csv('TFT_Platinum_MatchData.csv')
    df_champ = pd.read_csv('TFT_Champion_CurrentVersion.csv')
    df_item = pd.read_csv('TFT_Item_CurrentVersion.csv')
    
elif platform.system() == 'Mac':
    df_ch = pd.read_csv('TFT_Tier1_CH_MatchData.csv')
    df_gm = pd.read_csv('TFT_Tier2_GM_MatchData.csv')
    df_ma = pd.read_csv('TFT_Tier3_MA_MatchData.csv')
    df_di = pd.read_csv('TFT_Tier4_DI_MatchData.csv')
    df_pt = pd.read_csv('TFT_Tier5_PT_MatchData.csv')
    df_champ = pd.read_csv('TFT_Champion.csv')
    df_item = pd.read_csv('TFT_Item.csv')
```

```python
# ‚úÖ ÎèôÏùº Íµ¨Ï°∞Ïùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÎîïÏÖîÎÑàÎ¶¨
dfs = {
    'df_ch': df_ch,
    'df_gm': df_gm,
    'df_ma': df_ma,
    'df_di': df_di,
    'df_pt': df_pt
}

# ‚úÖ Í≥µÌÜµ Ìï®Ïàò
def count_empty_cases(df):
    cond_champion = df['champion'] == '{}'
    cond_comb     = df['combination'] == '{}'
    return {
        'champion_only': (cond_champion & ~cond_comb).sum(),
        'comb_only': (~cond_champion & cond_comb).sum(),
        'both_empty': (cond_champion & cond_comb).sum(),
        'total_empty': (cond_champion | cond_comb).sum()
    }

# ‚úÖ Í≤∞Í≥º Ï†ÄÏû•Ïö© Î¶¨Ïä§Ìä∏
results = []

# ‚úÖ Í∞úÎ≥Ñ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï∂úÎ†•
for name, d in dfs.items():
    counts = count_empty_cases(d)
    results.append(counts)
    print(f"\nüìä {name} Í≤∞Í≥º")
    print(f"1Ô∏è‚É£ championÎßå Îπà Ìñâ Í∞úÏàò      : {counts['champion_only']}")
    print(f"2Ô∏è‚É£ combinationÎßå Îπà Ìñâ Í∞úÏàò    : {counts['comb_only']}")
    print(f"3Ô∏è‚É£ Îëò Îã§ Îπà Ìñâ Í∞úÏàò           : {counts['both_empty']}")
    print(f"4Ô∏è‚É£ Ï¥ùÌï© (Îëò Ï§ë ÌïòÎÇòÎùºÎèÑ Îπà Ìñâ) : {counts['total_empty']}")

# ‚úÖ Ï†ÑÏ≤¥ Ï¥ùÌï© Í≥ÑÏÇ∞
total_sum = {key: sum(r[key] for r in results) for key in results[0].keys()}

print("\n===============================")
print("üìà Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï¥ùÌï©")
print(f"1Ô∏è‚É£ championÎßå Îπà Ìñâ Ï¥ùÌï©      : {total_sum['champion_only']}")
print(f"2Ô∏è‚É£ combinationÎßå Îπà Ìñâ Ï¥ùÌï©    : {total_sum['comb_only']}")
print(f"3Ô∏è‚É£ Îëò Îã§ Îπà Ìñâ Ï¥ùÌï©           : {total_sum['both_empty']}")
print(f"4Ô∏è‚É£ Ï¥ùÌï© (Îëò Ï§ë ÌïòÎÇòÎùºÎèÑ Îπà Ìñâ) : {total_sum['total_empty']}")
print("===============================")
```
