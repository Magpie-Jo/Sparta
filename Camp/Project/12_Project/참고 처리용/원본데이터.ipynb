```python
#Ï¥àÍ∏∞ ÏÑ§Ï†ïÎ∞è ÏãúÏä§ÌÖú ÎùºÏù¥Î∏åÎü¨Î¶¨
import platform
import warnings

# Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî ÎùºÏù¥Î∏åÎü¨Î¶¨
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta

warnings.filterwarnings('ignore')

print(platform.system())
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_colwidth', None)

# ÏãúÍ∞ÅÌôî OSÎ≥Ñ ÌïúÍ∏ÄÌè∞Ìä∏ ÏÑ§Ï†ï
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows Ìè∞Ìä∏ ÏÑ§Ï†ï
elif platform.system() == 'Mac':
    plt.rcParams['font.family'] = 'AppleGothic'  # Mac Ìè∞Ìä∏ ÏÑ§Ï†ï
```

```python
# Ìã∞Ïñ¥Î≥Ñ ÌååÏùºÍ≥º Ïù¥Î¶Ñ Îß§Ìïë
tier_files = {
    "Challenger": "TFT_Challenger_MatchData.csv",
    "GrandMaster": "TFT_GrandMaster_MatchData.csv",
    "Master": "TFT_Master_MatchData.csv",
    "Diamond": "TFT_Diamond_MatchData.csv",
    "Platinum": "TFT_Platinum_MatchData.csv"
}

# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Î≥ëÌï©
df_list = []
for tier, file in tier_files.items():
    df = pd.read_csv(file, encoding='utf-8')
    df["tier"] = tier
    print(f"[{tier}] Ï≤òÎ¶¨ Ï†Ñ Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞: {df.shape}")
    df_list.append(df)

# ÌÜµÌï©
df = pd.concat(df_list, ignore_index=True) 
# ignore_index = Í∏∞Ï°¥ Ïù∏Îç±Ïä§ Î≤ÑÎ¶¨Í≥† ÏÉà Îç∞Ïù¥ÌÑ∞Ïóê 0Î∂ÄÌÑ∞ ÏàúÏ∞®Ï†ÅÏúºÎ°ú Ïù∏Îç±Ïä§ Î∂ÄÏó¨
print("="*60)
print("ÌÜµÌï© ÌõÑ Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞:", df.shape)
```

```python
# ‚úÖ Ï°∞Í±¥ Ï†ïÏùò
cond1 = df['champion'] == '{}'
cond2 = df['combination'] == '{}'
cond3 = df['Ranked'] == 0

# ‚úÖ Í≤∞Í≥ºÎ≥Ñ Ïπ¥Ïö¥Ìä∏
result1 = (cond1 & ~cond2 & ~cond3).sum()   # Ï°∞Í±¥ 1Îßå
result2 = (~cond1 & cond2 & ~cond3).sum()   # Ï°∞Í±¥ 2Îßå
result3 = (~cond1 & ~cond2 & cond3).sum()   # Ï°∞Í±¥ 3Îßå
result4 = (cond1 & cond2 & ~cond3).sum()    # Ï°∞Í±¥ 1 + 2
result5 = (cond1 & ~cond2 & cond3).sum()    # Ï°∞Í±¥ 1 + 3
result6 = (~cond1 & cond2 & cond3).sum()    # Ï°∞Í±¥ 2 + 3
result7 = (cond1 & cond2 & cond3).sum()     # Ï°∞Í±¥ 1 + 2 + 3 (Î™®Îëê)

# ‚úÖ Ï¥ùÌï©
total = result1 + result2 + result3 + result4 + result5 + result6 + result7

# ‚úÖ Í≤∞Í≥º Ï∂úÎ†•
print("üìä Ï°∞Í±¥Î≥Ñ Ï°∞Ìï© Í≤∞Í≥º")
print(f"1Ô∏è‚É£ championÎßå Îπà ÎîïÏÖîÎÑàÎ¶¨          : {result1}")
print(f"2Ô∏è‚É£ combinationÎßå Îπà ÎîïÏÖîÎÑàÎ¶¨       : {result2}")
print(f"3Ô∏è‚É£ Ranked == 0Îßå Ìï¥Îãπ             : {result3}")
print(f"4Ô∏è‚É£ champion & combination        : {result4}")
print(f"5Ô∏è‚É£ champion & Ranked==0          : {result5}")
print(f"6Ô∏è‚É£ combination & Ranked==0       : {result6}")
print(f"7Ô∏è‚É£ Î™®Îì† Ï°∞Í±¥ ÎßåÏ°± (1+2+3)         : {result7}")
print("‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ")
print(f"Ï¥ùÌï© (Ï§ëÎ≥µ ÏóÜÏù¥ Ï†ÑÏ≤¥ Ï°∞Í±¥Ìï©)       : {total}")
```

```python
def check_str_length(df, col_name, top_n=20):
    """
    ‚úÖ ÌäπÏ†ï Ïª¨ÎüºÏùò Î¨∏ÏûêÏó¥ Í∏∏Ïù¥Î•º Í≥ÑÏÇ∞ÌïòÍ≥†, ÏßßÏùÄ ÏàúÏúºÎ°ú Ï†ïÎ†¨Ìï¥ÏÑú Î∞òÌôòÌïòÎäî Ìï®Ïàò
    
    Parameters
    ----------
    df : pd.DataFrame
        ÌôïÏù∏Ìï† Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ
    col_name : str
        Î¨∏ÏûêÏó¥ Í∏∏Ïù¥Î•º ÌôïÏù∏Ìï† Ïª¨Îüº Ïù¥Î¶Ñ
    top_n : int, optional
        ÏÉÅÏúÑ nÍ∞úÏùò ÏßßÏùÄ Îç∞Ïù¥ÌÑ∞Îßå ÎØ∏Î¶¨Î≥¥Í∏∞Ïö©ÏúºÎ°ú Ï∂úÎ†• (Í∏∞Î≥∏Í∞í 20)

    Returns
    -------
    pd.DataFrame
        Ìï¥Îãπ Ïª¨ÎüºÏùò Í≥†Ïú†Í∞íÍ≥º Í∏∏Ïù¥, Í∏∏Ïù¥Ïàú Ï†ïÎ†¨ Í≤∞Í≥º
    """
    # Í≥†Ïú†Í∞íÎßå Ï∂îÏ∂ú
    unique_vals = df[col_name].astype(str).unique()
    
    # DataFrame Î≥ÄÌôò
    result = pd.DataFrame(unique_vals, columns=[col_name])
    
    # Î¨∏ÏûêÏó¥ Í∏∏Ïù¥ Í≥ÑÏÇ∞
    result['length'] = result[col_name].astype(str).str.len()
    
    # ÏßßÏùÄ ÏàúÏúºÎ°ú Ï†ïÎ†¨
    result = result.sort_values(by='length', ascending=True).reset_index(drop=True)
    
    # Í≤∞Í≥º ÎØ∏Î¶¨Î≥¥Í∏∞
    print(f"‚úÖ '{col_name}' Ïª¨Îüº Î¨∏ÏûêÏó¥ Í∏∏Ïù¥ Î∂ÑÏÑù Í≤∞Í≥º (ÏßßÏùÄ Ïàú)")
    display(result.head(top_n))
    
    # Í≤∞Í≥º Ï†ÑÏ≤¥ Î∞òÌôò
    return result
```

```python
# Ïª¨Îüº Í∞í Ï≤¥ÌÅ¨
check_str_length(df, 'combination', top_n=50)
```
