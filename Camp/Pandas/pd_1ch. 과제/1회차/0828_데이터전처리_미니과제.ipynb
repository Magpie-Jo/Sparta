## 문제 7: 행 삭제(drop) 연습
bmi 값이 결측치(NaN) 이거나, bmi < 10 또는 bmi > 80인 행은 잘못된 데이터라고 가정합시다.  
이런 행들을 제거한 새로운 데이터프레임 df_clean을 만드세요. (원본 df는 그대로 둡니다.)  
제거 전/후 행 개수를 비교해서 출력해보세요.  

```python
#제거 목표 : BMI < 10, BMI > 80 , BMI = NaN 

cond_index = df[ (df['bmi'] > 80) | (df['bmi'] < 10) ].index

df_drop = df.drop(cond_index).dropna()

print(df_drop,'n/')

print ("원본데이터 개수 : " , df['bmi'].shape)         #원본데이터 인포
print ("원본데이터 개수 : " , df_drop['bmi'].shape )    #드롭데이터 인포
```
