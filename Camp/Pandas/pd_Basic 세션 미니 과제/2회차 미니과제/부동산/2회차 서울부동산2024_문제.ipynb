# 서울시 부동산 2024 (2024.csv)

## 문제 0 - 데이터 로딩
데이터는 다음 경로에서 읽어옵니다: `2024.csv`  
encoding="cp949" 옵션 권장

```python
import pandas as pd

df = pd.read_csv('2024.csv', encoding= 'cp949')
print(df.shape)

df.head()
```

### 문제 1 — 조건 필터링(AND)
- 자치구명이 '강남구' **이면서** 건물용도가 '오피스텔'인 행만 필터링하세요.

```python
df1 = df

cond_local = (df1['자치구명'] == '강남구')
cond_Build_use = (df1['건물용도'] == '오피스텔')

all_cond = cond_local & cond_Build_use

df1 = df1[all_cond]

df1.head()
```

### 문제 2 — 조건 필터링(OR)
- 층(`층`)이 10층 이상 **또는** 물건금액(만원)이 100000 이상인 행만 필터링하세요.

```python
df2 = df

cond_floar = (df['층'] <= 10)
cond_price = (df['물건금액(만원)']) <= 100000

all_cond2 = cond_floar | cond_price

df2 = df2[all_cond2]

df2.head()
```

### 문제 3 — 결측치 조건
- 취소일이 결측(`NaN`)인 행만 필터링하세요. (거래 취소가 없는 건)

```python
df3 = df

df3 = df3[df3['취소일'].isnull()]

print(f'취소일 결측치 개수 : {df3['취소일'].isnull().sum()}')
df3.head()
```

### 문제 4 — 특정 값 필터링
- 신고구분이 '직거래'인 행만 필터링하세요.

```python
df4 = df

cond_report = df4['신고구분'] == '직거래'

df4 = df4[cond_report]

df4.head()
```

### 문제 5 — 정렬
- 물건금액(만원) 기준 **내림차순** 정렬 후 상위 10행을 출력하세요.

```python
df5 = df

df5 = df5.sort_values(by='물건금액(만원)', ascending= False)

df5.head()
```

### 문제 6 — 다중 정렬
- 건축년도 **오름차순**, 건물면적(㎡) **내림차순**으로 다중 정렬하세요.

```python
df6 = df

df6 = df5.sort_values(by=['건축년도','건물면적(㎡)'], ascending= [True,False])

df5.head()
```

### 문제 7 — 열 제거
- 분석에 불필요하다고 가정하고 `권리구분` 열을 제거한 새로운 DataFrame을 만들고 컬럼 목록을 출력하세요.

```python
df7 = df
print(f'제거 전 shape : {df7.shape}')

df7 = df7.drop(columns='권리구분')

print(f'제거 후 shape : {df7.shape}')
df7.head()

```

### 문제 8 — 열 이름 변경
- `물건금액(만원)`→`거래금액_만원`, `건물면적(㎡)`→`건물면적_m2`, `토지면적(㎡)`→`토지면적_m2`로 변경하세요.

```python
df8 = df

df8 = df8.rename(columns={ 
     '물건금액(만원)' : '거래금액_만원' 
    ,'건물면적(㎡)'   : '건물면적_m2'
    ,'토지면적(㎡)'   : '토지면적_m2'
})

df8
```

### 문제 9 — 결측치 개수
- 각 열의 결측치 개수를 한 번에 확인하세요.

```python
df9 = df

df9_null = df9.isnull().sum()

df9_null
```

### 문제 10 — 결측치 채우기(수치형)
- 건축년도 결측치를 건축년도 **중앙값**으로 채운 새 DataFrame을 만들고, 채운 후 해당 열의 결측치 개수를 출력하세요.

```python
df10 = df
build_year = df10['건축년도']
print(f'결측치 대체 전 : {build_year.isnull().sum()}')

median_year = build_year.median()
print(f'대체 연도 : {median_year}')

build_year = build_year.fillna(median_year)
df10['건축년도'] = build_year

print(f'결측치 대체 후 : {build_year.isnull().sum()}')
```

### 문제 11 — 결측치 채우기(문자열)
- 건물명 결측치를 '미기재'로 채우고, 채운 후 해당 열의 결측치 개수를 출력하세요.

```python
df11 = df
build_name = df11['건물명']
print(f'결측치 대체 전 : {build_name.isnull().sum()}')

build_name = build_name.fillna('미기재')
print(f'결측치 대체 후 : {build_name.isnull().sum()}')
```

### 문제 12 — 중복 확인 및 제거(subset)
- 중복 기준을 `['자치구명','법정동명','계약일','건물면적(㎡)','물건금액(만원)']`으로 하여 중복 개수를 확인하고, 이를 기준으로 중복을 제거한 뒤 제거 전/후 행 수를 비교하세요.

```python
df12 = df
print(f'중복치 제거 전 : {df12.duplicated().sum()}')

columns = ['자치구명','법정동명','계약일','건물면적(㎡)','물건금액(만원)']

df12 = df12.drop_duplicates(subset=columns, keep=False)

print(f'중복치 제거 후 : {df12.duplicated().sum()}')
```
