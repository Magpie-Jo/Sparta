# 서울시 부동산 2024 데이터 문제 세트 (Pandas)

```python
# ✅ 기본 설정 (수정 금지)
import pandas as pd

df = pd.read_csv("2024.csv", encoding="cp949")
print(df.shape)
df.head()
```

---

### 문제 1 — 시간: `계약일`을 datetime으로 변환
• `계약일`(예: 20241014)을 문자열로 바꾼 뒤 `format='%Y%m%d'`로 `pd.to_datetime` 하여 `계약일_dt` 열을 만드세요.

```python
df1 = df

# 컬럼 추출
계약일_dt = df1['계약일']

# 형변환(포맷팅)
df1['계약일_dt'] = pd.to_datetime(계약일_dt, format = '%Y%m%d' , errors='coerce')

# 해당 컬럼 데이터 프레임 할당
dt =  df1[['계약일','계약일_dt']]

# 출력 확인
dt.head()
```

---

### 문제 2 — 시간: 계약 연/월/일 추출
• `계약일_dt`에서 연(`year`), 월(`month`), 일(`day`)을 각각 `계약연`, `계약월`, `계약일자` 열로 추출하세요. (`.dt`)

```python
dt_year = dt['계약일_dt'].dt.year
dt_month = dt['계약일_dt'].dt.month
dt_day = dt['계약일_dt'].dt.day


dt['계약연'] = dt_year
dt['계약월'] = dt_month
dt['계약일자'] = dt_day

dt[['계약연','계약월','계약일자']].head()
```

### 문제 3 — 그룹화: 자치구별 평균 거래금액
• `자치구명`별 `물건금액(만원)`의 평균을 구하세요. (결과를 내림차순 정렬해도 무방)

```python
df3 = df

# 자치구명 별 물건금액 평균 산출(시리즈)
local_mean = (
    df3.groupby('자치구명')['물건금액(만원)']
    .mean()
    .astype(int)
    .reset_index())

# 금액 내림차순 정렬
local_mean = (
    local_mean.sort_values(
        by = '물건금액(만원)' 
        ,ascending=False))

# 결과 출력
local_mean
```

### 문제 4 — 그룹화: (자치구, 건물용도)별 건수와 평균금액
• `['자치구명','건물용도']`로 그룹화하여 건수(`.size()` 또는 `.count()`)와 평균금액(`mean`)을 함께 구하세요. (`.agg`)

```python
df4 = df

df4_agg = df4.groupby(['자치구명','건물용도'])['물건금액(만원)'].agg(['size','count','mean']).astype(int)

df4_agg.head(10)
```

### 문제 7 — apply/lambda: ㎡당 단가(만원) 계산
• 각 행에 대해 `단가_만원_㎡ = 물건금액(만원) / 건물면적(㎡)` 열을 `DataFrame.apply`(axis=1)로 계산하세요.  
• 계산 후, 자치구별 평균 단가를 구하세요.

```python
df7 = df

df7['단가_만원_㎡'] = df.apply(
    lambda row: row['물건금액(만원)'] / row['건물면적(㎡)'] ,
    axis=1
)


borough_mean_unit_price = df7.groupby('자치구명')[['단가_만원_㎡']].mean()

borough_mean_unit_price.sort_values(by = '단가_만원_㎡', ascending= False).astype(int)
```

### 문제 8 — map: 신고구분 설명 라벨 추가
• 매핑 사전 예시: `{{'중개거래': '중개사 통해 거래', '직거래': '직접 거래'}}` → `신고구분_설명` 열을 추가하세요.

### 문제 9 — Merge: 자치구 권역 라벨 병합
• 다음과 같은 매핑 DataFrame을 생성하세요 (예시는 일부 자치구만 포함):  
  `zone_df = pd.DataFrame({{'자치구명':['강서구','영등포구','성동구','서대문구','은평구','동대문구','금천구'], '권역':['서남권','서남권','동북권','서북권','서북권','동북권','서남권']}})`  


• `df`와 `zone_df`를 `자치구명` 기준으로 **merge**하여 `권역` 열을 추가하세요. (merge를 선택해서 수행하세요 -> 선택한 merge에는 이유가 있어야 함)

```python
df9 = df.copy(deep=True)

zone_df = pd.DataFrame({
    '자치구명':
    ['강서구','영등포구','성동구','서대문구','은평구','동대문구','금천구'], 
      
    '권역':
    ['서남권','서남권','동북권','서북권','서북권','동북권','서남권']
})

df9_district = pd.merge(df9,zone_df , on = '자치구명', how= 'inner')
df9_district['권역'].value_counts()
```

### 문제 10 — Concat: 짝수/홀수 인덱스 분할 후 복원
• `df_even = df.iloc[::2]`, `df_odd = df.iloc[1::2]`로 분할 후 `pd.concat`으로 다시 합치고, 원본 행 수와 같은지 확인하세요.

```python
df10 = df.copy(deep=True)
print(f'전체 shape : {df10.shape}\n')

df10_even = df10.iloc[::2]
df10_odd = df10.iloc[1::2]

print(f'짝수 : {df10_even.shape}')
print(f'홀수 : {df10_odd.shape}\n')

df10_concat = pd.concat([df10_even,df10_odd], axis=0)

print(f'통합 : {df10_concat.shape}')
```

### 문제 11 — 시간: 취소일과의 차이(일수) 계산
 • `취소일`을 `pd.to_datetime(..., format='%Y%m%d', errors='coerce')`로 변환한 `취소일_dt` 열을 만들고,  
• `취소일_dt - 계약일_dt`의 일수 차(`days_to_cancel`)를 계산하세요.

```python
# TODO: 여기에서 풀이 코드를 작성하세요.
# 먼저 취소일 결측치를 없애고 계산 -> 그렇지 않으면 NaT(Not a Time) 값이 다수 나옴.
df11 = df.copy(deep=True)

df11 = df11.dropna(subset=['취소일'])

df11['취소일_dt'] = pd.to_datetime(df11['취소일'], format='%Y%m%d', errors='coerce')

df11_diff = df11['취소일_dt'] - df11['계약일_dt']

df11['days_to_canel'] = df11_diff

df11[['계약일_dt', '취소일_dt', 'days_to_canel']].head(10)
```
