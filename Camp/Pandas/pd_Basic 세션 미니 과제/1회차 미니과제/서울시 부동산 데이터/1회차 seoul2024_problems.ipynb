```python
fill_columns = ['자치구명','지번구분명','지번구분','본번','부번','건물명','토지면적(㎡)','층','건축년도'] #결측 존재 컬럼 지정

df2.loc[[21265],['건물명']] = '연립다세대' #단일 값으로 별도 지정
df2.loc[[73896],['본번']] = 0 
df2.loc[[73896],['부번']] = 0

fill_NaN = {
    '자치구명' : '강북구',
    '지번구분명': '미상',
    '지번구분' : 0.0,
    '본번' : 0,
    '부번' : 0,
    '건물명' : '단독다가구',
    '토지면적(㎡)' : '미측정',
    '층' : 1.0,
    '건축년도' : 0000
}                                      #결측치 수정용 딕셔너리 생성

df2[fill_columns] = df2[fill_columns].fillna(fill_NaN) # 결측지 수정 적용
```
