# 실습 1 — 매출(revenue) 붙이고 총합 구하기

## 데이터

```python
rows_p1 = [
    {'price': 1000, 'qty': 2},
    {'price':  700, 'qty': 3},
    {'price':  300, 'qty': 4},
]
```

## Before: 함수화하지 않은 코드

```python
# 각 행에 revenue = price * qty 추가
rows_p1_new = []
for r in rows_p1:
    price = r['price']
    qty = r['qty']
    revenue = price * qty

    new_row = {}
    new_row['price'] = price
    new_row['qty'] = qty
    new_row['revenue'] = revenue
    rows_p1_new.append(new_row)

# 총 매출 합계 계산
total = 0
for r in rows_p1_new:
    total = total + r['revenue']

print(rows_p1_new)  # 확인용
print("총합:", total)  # 기대: 5300
```

## 요구사항 (함수화)
1. add_revenue(row) — 한 행에 revenue를 붙여 새 행으로 반환
2. add_revenue_all(rows) — 리스트 전체에 add_revenue을 적용해 새 리스트 반환
3. sum_revenue(rows_with_rev) — revenue를 모두 더해 총합 반환

마크다운
1. `add_revenue(row)`  
   - 입력: `{'price': 가격, 'qty': 수량}` 형태의 딕셔너리 1개  
   - 처리: `revenue = price * qty` 계산  
   - 출력: `{'price', 'qty', 'revenue'}`가 들어 있는 새 딕셔너리

2. `add_revenue_all(rows)`  
   - 입력: 여러 행이 들어 있는 리스트  
   - 처리: 각 행에 `add_revenue`를 적용  
   - 출력: 각 행에 `revenue`가 붙어 있는 새 리스트

3. `sum_revenue(rows_with_rev)`  
   - 입력: `revenue`가 포함된 행 리스트  
   - 처리: 모든 행의 `revenue`를 더함  
   - 출력: 총 매출(정수)

> 포인트  
> - 같은 계산을 여기저기 복붙하지 않고, **함수 하나에만 정의**해 두면 실수와 중복이 줄어듭니다.  
> - 앞으로 전처리, 시각화, 통계 분석, 머신러닝에서도 이런 식으로 **단계별로 함수화**하면 코드가 훨씬 읽기 쉬워집니다.

## 실습 2 — 특정 도시만 필터하고 합계 구하기

## 데이터

```python
rows_p2 = [
    {'city': 'Seoul', 'revenue': 2400},
    {'city': 'Busan', 'revenue': 2100},
    {'city': 'Seoul', 'revenue':  800},
]
target_city = 'Seoul'
```

## Before: 함수화하지 않은 코드

```python
# target_city만 남기기
filtered = []
for r in rows_p2:
    if r['city'] == target_city:
        filtered.append(r)

# revenue 합계
total = 0
for r in filtered:
    total = total + r['revenue']

print(filtered)  # 확인용
print("합계:", total)  # 기대: 3200
```

## 요구사항(함수화)
1. filter_by_city(rows, city) — city가 일치하는 행만 새 리스트로 반환
2. sum_revenue(rows) — 리스트의 revenue를 더해 합계 반환

마크다운

1. `filter_by_city(rows, city)`  
   - 입력: `rows` = `{'city': ..., 'revenue': ...}` 형태의 딕셔너리 리스트  
   - 처리: `city`가 지정한 값과 같은 행만 골라 새로운 리스트로 모음  
   - 출력: 해당 도시 행들만 담고 있는 새 리스트

2. `sum_revenue(rows)`  
   - 입력: `rows` = `{'revenue': ...}`가 들어 있는 딕셔너리 리스트  
   - 처리: 모든 `revenue` 값을 더함  
   - 출력: 매출 합계(정수)

> 포인트
> - “필터링”과 “합계 계산”을 분리해 두면  
>   - 다른 도시에도 그대로 재사용할 수 있고  
>   - 나중에 조건이 바뀌어도 **필터 함수만 수정**하면 돼서 유지보수가 쉬워집니다.

```python
def sum_list(num_list):
    '''
    숫자 리스트의 합 계산을 반환 하는 함수
    '''
    return sum(num_list) # num_list 전체합


def avg_list(num_list):
    '''
    숫자 리스트의 평균 계산을 반환 하는 함수
    '''
    return sum_list(num_list)/len(num_list) # num_list 전체합 / num_list 개수 


def print_sum_and_avg(numbers):
    '''
    숫자 리스트의 합과 평균을 출력하는 함수
    '''
    sum_nums = sum_list(numbers)
    avg_nums = avg_list(numbers)

    print(f"합 : {sum_nums}\n평균 : {avg_nums}")


numbers = [10, 20, 30, 40]
print_sum_and_avg(numbers)
```

```python
scores = [80, 55, 90, 70]

# 60점 이상이면 합격
pass_count = 0
for s in scores:
    if s >= 60:
        pass_count = pass_count + 1

print("합격자 수:", pass_count)
```

```python
def is_pass(scores):
    if scores >= 60:
        return True
    else:
        return False

def count_pass(scores):
    count = 0
    for score in scores:
        if is_pass(score):
            count += 1
    return count

count_pass(scores)


```
