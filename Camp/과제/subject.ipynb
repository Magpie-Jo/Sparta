```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

total_score = (sp['math score']>=80) & (sp['reading score']>=80) & (sp['writing score']>=80)

score_sp = sp[total_score]
vc_sp = score_sp['gender'].value_counts()

vc_sp


vc_sp.plot.bar(width=0.9,color=['deeppink','deepskyblue'])
plt.title("High Achievers(>=80 in all) by Gender")
plt.xticks(rotation=0)
plt.xlabel('gender',size = 18)
plt.ylabel('count',size= 18)

print(vc_sp)
show_sp = plt.show()
```

문제 5

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

cond1 = (sp['lunch']=='standard')
cond2 = (sp['test preparation course']=='completed')

cond = cond1 & cond2
f_df = sp[cond]

avg = f_df.loc[ : , ['math score', 'reading score', 'writing score']].mean().round(1)

avg.plot.bar(
            color = ['cornflowerblue','deepskyblue','aqua']
            )

plt.title ('Avg scores (prep=completed & lunch=standard)')
plt.ylim (0,100)
plt.ylabel ('score')
print(avg)

visual_avg=plt.show()


```

문제 6

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

par_lv_edu = sp
grp_sp = (
    par_lv_edu.groupby('parental level of education') ['math score']
    .mean()
    .round(1)
)
srt_sp = grp_sp.sort_values(ascending=False)
df = srt_sp.reset_index()

x = srt_sp.index
y = srt_sp.values

plt.scatter(x,y)

plt.title('Mean Math Score by Parental Education(scatter)')

plt.xlabel("parental education")
plt.xticks(rotation = 30, ha='right')

plt.ylabel("mean math score")
plt.ylim(0,100)

visual_scatter = plt.show()


```
