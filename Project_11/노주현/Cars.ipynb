```python
## 사용 패키지 임포트
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib.font_manager as fm
```

```python
import pandas as pd
# CSV 데이터 불러오기
cars = pd.read_csv('cars.csv')

# 데이터 프레임 생성 
cars = pd.DataFrame(cars)
```

```python
# 데이터 프레임 불러오기
cars
```

```python
Cars_info = cars.info()                                         # 테이블 정보 (인덱스 , 컬럼, 컬럼별 데이터수, 컬럼타입)
Cars_shape = cars.shape                                         # 테이블 쉐이프 (데이터수 , 컬럼수)}

Cars_shape_df = pd.DataFrame({'데이터 개수' : [Cars_shape[0]],  # 데이터 프레임화
                                '컬럼수' : [Cars_shape[1]]
                             }
                            )

print(Cars_info, '\n')
Cars_shape_df
```

```python
# 결측치 구하기
cars_isnull = cars.isnull().sum()       #결측치 확인
cars_isnull_df = pd.DataFrame(
    cars_isnull , columns=['결측치 개수']
                            )           #시리즈 데이터 테이블화

cars_isnull_df                          #결과물
```

```python
#컬럼별 기초 통계 
cars_dscr = cars.describe().round(2)                                # 기초 통계량 산출
cars_dscr_df = pd.DataFrame(cars_dscr , index = cars_dscr.index)    # 결과물 데이터프레임 전환

cars_dscr_df                                                        # 컬럼별 통계치 확인 
```

```python
# 중복치 제거
cars_dup = cars.drop_duplicates()   # 중복 제거
cars_complete = cars_dup            # 결과물 데이터 저장

before = cars.shape                 # 처리 전 쉐이프
after = cars_dup.shape              # 처리 후 쉐이프

dup_shape_df = pd.DataFrame({"데이터 개수": [before[0], after[0]],
                            "컬럼 개수": [before[1], after[1]]
                            }
                        , index=["중복 제거 전", "중복 제거 후"]
                        )           # 쉐이프 데이터 프레임화

cars_dup                            # 중복 제거 결과물
dup_shape_df                        # 비교 출력
```

```python
cars_upper = cars_complete                              # 이전 데이터 저장

cars_upper['MAKE'] = cars_upper['MAKE'].str.upper()     # 제조사 대문자 통일 변환
cars_complete = cars_upper                              # 결과물 데이터 저장

before= cars['MAKE'].value_counts()                     # 원본데이터
after = cars_upper['MAKE'].value_counts()               # 변환 후 데이터

print(before,'\n ---------------------------------')    # 변환 전 제조사 벨류 카운트
print(cars_upper['MAKE'].value_counts())                # 변환 후 제조사 벨류 카운트
```
