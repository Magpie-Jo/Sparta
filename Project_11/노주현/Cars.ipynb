```python
city_km_l = (100/cars['CITY (L/100 km)']).round(2)
HWY_km_l  = (100/cars['HWY (L/100 km)']).round(2)
COMB_km_l = (100/cars['COMB (L/100 km)']).round(2)



onry_fuel = pd.DataFrame({
    "fuel": cars["FUEL"],
    "city (km/l)": city_km_l,
    "hwy (km/l)": HWY_km_l,
    "comb (km/l)": COMB_km_l,
    "emissions": cars["EMISSIONS"]
}, index=cars.index)

```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic' # For Windows
# plt.rcParams['font.family'] = 'AppleGothic' # For MacOS
%matplotlib inline

# 연료 이름과 에너지 밀도
prop_default= 33.5
e = 23
x = 33.5
n = 21
d = 38.6
z = 34.2
fuels = [ 'x (휘발유)', 'z (고급휘발유)', 'd (디젤)', 'e (E85)','n (천연가스)']
energy_density = [33.5, 34.2, 38.6, 23, 21]  # MJ/L


fuel_efficiency = [prop_default/33.5, prop_default/34.2, prop_default/38.6, prop_default/23, prop_default/21]

# 막대그래프 생성
plt.figure(figsize=(8,6),dpi=150)
plt.bar(fuels, energy_density, color=['Yellow', 'red', 'darkgreen', 'mediumspringgreen','dodgerblue'])
plt.ylabel('에너지 밀도 (MJ/L)')
plt.title('연료별 에너지 밀도 비교')
plt.show()

# 막대그래프 생성
plt.figure(figsize=(8,6),dpi=150)
prop_bar=plt.bar(fuels, fuel_efficiency, color=['Yellow', 'red', 'darkgreen', 'mediumspringgreen','dodgerblue'])
plt.ylabel('energy prop')
plt.title('상대적 에너지 소비 지표 (일반휘발유 : 1)')
plt.show(prop_bar)
```
